```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np

# Step 2: Load the CSV file into a DataFrame (replace 'file_path' with the actual path to your CSV file)
data = pd.read_csv('pearson.csv')

# Step 3: Display the first few rows of the data to understand its structure
print(data.head())

# Step 4: Calculate Pearson's correlation matrix for the features
correlation_matrix = data.corr(method='pearson')

# Step 5: Display the correlation matrix
print(correlation_matrix)

# Optional: To visualize the correlation matrix, use a heatmap (requires seaborn and matplotlib)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Pearson's Correlation Matrix")
plt.show()
```
